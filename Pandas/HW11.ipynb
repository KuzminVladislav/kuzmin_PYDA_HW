Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
	- оценка 2 и меньше - низкий рейтинг
	- оценка 4 и меньше - средний рейтинг
	- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class


```python
movie = pd.read_csv ('movies.csv') # данные с фильмами
rate = pd.read_csv ('ratings.csv') # данные с рейтингами
```

```python
rezult = pd.merge(left=movie,right=rate,on="movieId") # Объединяем данные по movieId
rezult.head()
```

```python
def rate (rating):
    if rating <= 2:
        return "Low"
    elif rating <= 4:
        return "Middle"
    elif rating > 4:
        return "High"
```

```python
rezult['class'] = rezult['rating'].apply(lambda x: rate (x))
```

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим 'undefined'.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv ('keywords.csv') # данные с фильмами
keywords.head()
```

```python
def regions (key):
    if ('москва' or 'тула' or 'ярославль') in key.keyword:
        return 'Центр'
    elif ('петербург'or 'псков'or 'мурманск') in key.keyword:
         return 'Северо-Запад'
    elif ('владивосток'or 'сахалин'or 'хабаровск') in key.keyword:
         return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
# keywords['region'] = keywords['keyword'].apply(lambda x: regions (x))
# keywords.head() # Аналогичное решение через подставление столбца
```

```python
keywords['region'] = keywords.apply(regions, axis=1) # добавление столбца через подставление строки
keywords.head(10)
```

```python
filtered_ = keywords[keywords['region'].isin (['Центр','Северо-Запад','Дальний Восток'])] 
filtered_.head(10) #Проверка
```
